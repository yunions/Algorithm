## Bubble sort

```python
sort =  [3,5,6,7,10,21,31]
```

```python
# [Do it! 실습 6-1] 버블 정렬 알고리즘 구현하기
# 수행횟수 표시하기
from typing import MutableSequence

def bubble_sort(a: MutableSequence) -> None:
    """버블 정렬"""
    n = len(a)
    for i in range(n - 1):                       # 맨 앞의 index=0부터 index=n-1까지 수행할 것
        for j in range(n - 1, i, -1):            # 맨 끝, index = n-1의 데이터부터 차례대로 단순 비교 시작
            if a[j - 1] > a[j]:
                a[j - 1], a[j] = a[j], a[j - 1]   # 2개의 데이터를 swap

if __name__ == '__main__':

    bubble_sort(sort)  # 배열 x를 버블 정렬

    print('오름차순으로 정렬했습니다.')
    for i in range(len(sort)):
        print(f'x[{i}] = {sort[i]}')
```

## Insertion

```python
from typing import MutableSequence

insertion = [3,4,1,9,7,5]

def insertion_sort(a: MutableSequence) -> None:
    """단순 삽입 정렬"""
    n = len(a)
    for i in range(1, n):
        j = i # j = index 1부터 시작
        tmp = a[i] #교환하기 위한 값 일시적 저장
        while j > 0 and a[j - 1] > tmp:
            a[j] = a[j - 1]
            j -= 1
        a[j] = tmp

if __name__ == '__main__':
    insertion_sort(insertion)  # 배열 x를 단순 삽입 정렬

    print('오름차순으로 정렬했습니다.')
    for i in range(len(insertion)):
        print(f'x[{i}] = {insertion[i]}')
```

```python
# [Do it! 실습 6C-1] 이진 삽입 정렬 알고리즘 구현하기

from typing import MutableSequence

insertion = [3,4,1,9,7,5]

def binary_insertion_sort(a: MutableSequence) -> None:
    """이진 삽입 정렬"""
    n = len(a)
    for i in range(1, n):
        key = a[i]
        pl = 0      # 검색 범위의 맨 앞 원소 인덱스
        pr = i - 1  # 검색 범위의 맨 끝 원소 인덱스

        while True:
            pc = (pl + pr) // 2  # 검색 범위의 중앙 원소 인덱스
            if a[pc] == key:     # 검색 성공
                break
            elif a[pc] < key:
                pl = pc + 1
            else:
                pr = pc - 1
            if pl > pr:
                break
    
        pd = pc + 1 if pl <= pr else pr + 1  # 삽입할 위치의 인덱스

        for j in range(i, pd, -1):
            a[j] = a[j - 1] #뒤로 넘기기
        a[pd] = key

if __name__ == "__main__":
    binary_insertion_sort(insertion)  # 배열 x를 이진 삽입 정렬

    print('오름차순으로 정렬했습니다.')
    for i in range(len(insertion)):
        print(f'x[{i}] = {insertion[i]}')
```

## Quick

```python
from typing import MutableSequence

def qsort(a: MutableSequence, left: int, right: int) -> None:
    """a[left] ~ a[right]를 퀵 정렬(배열을 나누는 과정 출력)"""
    pl = left                   # 왼쪽 index
    pr = right                  # 오른쪽 index
    x = a[(left + right) // 2]  # 피벗(가운데 원소)

    print(f'a[{left}] ~ a[{right}]: ', *a[left : right + 1])  # 새로 추가된 부분

    while pl <= pr:                     
        while a[pl] < x: pl += 1 #pl의 데이터가 x보다 작은 경우 pass! pl의 데이터가 x보다 큰 경우에 stop
        while a[pr] > x: pr -= 1 #pr의 데이터가 x보다 큰 경우 pass! pl의 데이터가 x보다 작은 경우에 stop
        # 위 두 while문을 나오는 경우 = pl index의 데이터가 x보다 큰 경우
        if pl <= pr:
            a[pl], a[pr] = a[pr], a[pl]
            pl += 1
            pr -= 1

    if left < pr: qsort(a, left, pr)   
    if pl < right: qsort(a, pl, right)

def quick_sort(a: MutableSequence) -> None:
    """퀵 정렬"""
    qsort(a, 0, len(a) - 1)

if __name__ == '__main__':
    print('퀵 정렬을 수행합니다(배열을 나누는 과정 출력).')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num    # 원소 수가 num인 배열을 생성

    for i in range(num):
        x[i] = int(input(f'x[{i}]: '))

    quick_sort(x)       # 배열 x를 퀵 정렬

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
```
